```python

response = requests.get("https://www.cbssports.com/college-basketball/teams/ALBANY/albany-great-danes/roster/")

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, "html.parser")

player_rows = soup.find_all("tr")

# Loop through each player row and gather information
for row in player_rows:
    player_data = {}
    player_columns = row.find_all("td")

    if player_columns:
        # Get the player name
        player_data["Number"] = player_columns[0].text.strip()
        
        # Get the player position
        player_data["Name"] = player_columns[1].text.strip()

        # Get the player height
        player_data["Position"] = player_columns[2].text.strip()

        # Get the player weight
        player_data["Height"] = player_columns[3].text.strip()

        # Get the player year
        player_data["Weight"] = player_columns[4].text.strip()
        
        # Get the players Year
        player_data["Year"] = player_columns[5].text.strip()
      
     # Get the players Hometown
        player_data["Hometown"] = player_columns[6].text.strip()
    
    # Print the gathered player information
        print(player_data)
```
