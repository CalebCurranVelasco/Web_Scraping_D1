```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

options =Options()
options.binary_location = r'C:\Program Files\Mozilla Firefox\firefox.exe'
driver = webdriver.Firefox(executable_path=r'C:\WebDrivers\geckodriver.exe', options=options)
url = f"https://www.cbssports.com/college-basketball/teams/"
driver.get(url)
```

```python
roster_links = driver.execute_script("return Array.from(document.querySelectorAll('a[href*=roster]')).map(el => el.href);")
```

```python
print(roster_links)
```

```python
def Team_info(teamUrl):

    response = requests.get(teamUrl)
    teams_dict = {}
    players_dict = {}
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    player_rows = soup.find_all("tr")
    player_database = []
    
    # find team name
    
    team_name = soup.find(id = "PageTitle-header")
    player_name = 0
    
    
    # Loop through each player row and gather information
    for row in player_rows:
        player_data = {}
        player_columns = row.find_all("td")

        if player_columns:
            # Get the player name
            player_data["Number"] = player_columns[0].text.strip()
            player_name = player_columns[1].text.strip()
            # Get the player position
            player_data["Name"] = player_columns[1].text.strip()

            # Get the player height
            player_data["Position"] = player_columns[2].text.strip()

            # Get the player weight
            player_data["Height"] = player_columns[3].text.strip()

            # Get the player year
            player_data["Weight"] = player_columns[4].text.strip()

            # Get the players Year
            player_data["Year"] = player_columns[5].text.strip()

         # Get the players Hometown
            player_data["Hometown"] = player_columns[6].text.strip()
            
        players_dict[player_name] = player_data
        
        teams_dict[team_name] = players_dict
    
    return(teams_dict)
```

```python
linkage = roster_links[:100]
Teams = {}
number = 0
for link in linkage:
    number += 1
    Teams[number] = Team_info(link)
```
